```python
def import_json_to_mongo(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
        if isinstance(data, list):
            collection.insert_many(data)
        else:
            collection.insert_one(data)
    print(f"Data from {file_path} imported into MongoDB.")
```

```python
# Iterate through all files in the folder
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    
    # Only process files 
    if os.path.isfile(file_path):
        # Handle JSON files
        if filename.endswith('.json'):
            import_json_to_mongo(file_path)
        
print("All files have been processed and imported into MongoDB.")
```

### Distribution of average prices

```python
import re
import matplotlib.pyplot as plt

documents = list(db.reviews.find())

average_prices = [] 

def parse_price(price_str):
    price_str = price_str.replace("$", "").replace("*", "").strip() # Remove symbols
    
    # Check if it's a range of price
    if " - " in price_str:
        low, high = map(float, price_str.split(" - "))
        return (low + high) / 2  #Average of the two values
    else:
        return float(price_str)  # Convert into float


for doc in documents:
    price_str = doc.get("HotelInfo", {}).get("Price")
    if price_str: 
        try:
            average_price = parse_price(price_str)
            average_prices.append(average_price)
        except ValueError:
            continue  

plt.figure(figsize=(10, 6))
plt.hist(average_prices, bins=20, color='green', edgecolor='black')
plt.title("Distribution of average price")
plt.xlabel("Average price ($)")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.savefig("average_price_distribution.png")
plt.show()
```

## Text Pre-Processing

Create a function to remove punctuation, stopwords, numbers, repeated characters and lemmatize

```python
import nltk
import re
from nltk.stem import WordNetLemmatizer
stop_words = nltk.corpus.stopwords.words('english')
from nltk.tokenize import WordPunctTokenizer
import string
wnl = WordNetLemmatizer()


def preprocessing(text):
  punctuationfree = "".join([i for i in text if i not in string.punctuation])
  nonumbers = re.sub(r'\d+', '', punctuationfree)

  pattern_alpha = re.compile(r"([A-Za-z])\1{2,}", re.DOTALL)
  formatted_text = pattern_alpha.sub(r"\1\1", nonumbers)
  pattern_punct = re.compile(r'([!$?])\1{1,}')
  combined_formatted = pattern_punct.sub(r"\1", formatted_text)


  tokenized_text = WordPunctTokenizer().tokenize(combined_formatted)
  tokenized_text_without_stopwords = [w for w in tokenized_text if not w.lower() in stop_words]

  tokens = [wnl.lemmatize(w) for w in tokenized_text_without_stopwords]
  return tokens
```

```python
df['tokenized'] = df['Text']
df['tokenized'] = df['tokenized'].apply(lambda x: preprocessing(x))
```

## Vectorization

## Classifers training and evaluation

We train different classifiers and evaluate them to find the best method of identifying Positive and Negative reviews. Inp articular we trained those classifiers:

- Random Forest
- Decision Tree
- Naive Bayes
- Support Vector Machine

### Random Forest

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train_vec, y_train)
y_pred_RF = model.predict(X_test_vec)

print("Random Forest:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_RF)}")
print(classification_report(y_test, y_pred_RF))
```

### Decision Tree

```python
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train_vec, y_train)
y_pred_DT = model.predict(X_test_vec)

print("Decision Tree:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_DT)}")
print(classification_report(y_test, y_pred_DT))
```

### Naive - Bayes

```python
model = MultinomialNB()
model.fit(X_train_vec, y_train)
y_pred_NB = model.predict(X_test_vec)

print("Naive Bayes:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_NB)}")
print(classification_report(y_test, y_pred_NB))
```

### Support Vector Machine

```python
model = SVC(probability=True)
model.fit(X_train_vec,y_train)
y_pred_SVC = model.predict(X_test_vec)

print("Support Vector Machine:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_SVC)}")
print(classification_report(y_test, y_pred_SVC))
```

```python
from sklearn.metrics import RocCurveDisplay
import matplotlib.pyplot as plt
```

```python
RocCurveDisplay.from_estimator(model, X_test_vec, y_test, color='green')
plt.savefig('ROC_BI.png', format='png')
plt.show()
```

In terms of *accuracy* all the classifiers provide a good performance, with the Support Vector Machine being the better overall with a accuracy score of 93%. As we're interested in identifying the Negative sentiment reviews we will give more importance to the *Recall* metric, where the Support Vector Machine is the best by far. Other than that, it is also the better classifier in terms of F-1 score, providing a good comprmise between *Precision* and *Recall*. So, we decided to use the SVC as the model for the prediction.

For a computational capacity problem, we decided to perform a global explaination using a sub-set of the data (300 instances) and selecting only 30 features in the vectorization

```python
import shap
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC

# Example data (replace with your actual dataset)
X_train_sample = X_train.sample(n=300, random_state=42)
y_train_sample = y_train[X_train_sample.index]

# Vectorizer with reduced features
vectorizer_reduced = TfidfVectorizer(stop_words=None, lowercase=True, max_features=30)
X_train_vec_reduced = vectorizer_reduced.fit_transform(X_train_sample)
X_test_vec_reduced = vectorizer_reduced.transform(X_test)

# Fit the model
model = SVC(probability=True, random_state=42)
model.fit(X_train_vec_reduced, y_train_sample)


```

```python
# Use SHAP KernelExplainer
explainer = shap.KernelExplainer(model.predict_proba, shap.sample(X_train_vec_reduced, 100))
shap_values = explainer.shap_values(X_train_vec_reduced)  # Explain the first test observation
```
